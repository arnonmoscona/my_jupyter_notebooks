```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels as sm

points = np.arange(-5,5,0.01) # 1000 equally spaced points
xs, ys = np.meshgrid(points, points)

z = np.sqrt(xs**2+ys**2)

# plot it
plt.imshow(z, cmap=plt.cm.rainbow)
plt.colorbar()
plt.title("Image of $\sqrt{x^2 + y^2}$ for a grid of values")
```

```python
# using np.where

xarr = np.array(range(11, 16))/10
yarr = np.array(range(21, 26))/10
cond = np.array([True, True, True, True, False])

# Array tertiary expression: choose from xarr or yarr based on cond
result = np.where(xarr, yarr, cond)
# 1s for True
result = np.where(cond, 1, 0)
# => array([1, 1, 1, 1, 0])

# pick from xarr when > 1.2 or zero:
result = np.where(xarr > 1.2, xarr, 0)
# => array([ 0. ,  0. ,  1.3,  1.4,  1.5])

result
```

```python
# basic stat functions

arr = np.random.randn(5,4)  # 5x4 matrix of normally distributed values
(arr.mean(), np.mean(arr), arr.sum())
# or along an axis:
arr.mean(axis=1)
# Normalize
normalized = (arr - arr.mean())/arr.std()
(normalized.mean(), normalized.std())  # very close to (0.0, 1.0)
# cumulative sum of absolutes across rows
abs(arr).cumsum(axis=1)
# or shorter:
abs(arr).cumsum(1)

# We can count as booleans are cooerced into 1 or 0
# So count positive numbers in the matrix along the columns, and over the whole array:
{'arr': arr, 'positives': arr>0, 'column positive counts': sum(arr>0), 'count all positives': sum(sum(arr>0))}
```

```python
import numpy as np
from numpy.linalg import inv, pinv, eig

diagonal_twos = np.identity(3) * 2
# regular matrix inversion:
inv(diagonal_twos)
# pseudo-invert:
pinv(diagonal_twos)  # same thing in this case
# and getting back identity:
twos = diagonal_twos
twos.dot(inv(twos))

eig(twos)
```
